```python
class Queue:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def enqueue(self, item):
        self.items.append(item)

    def dequeue(self):
        if not self.is_empty():
            return self.items.pop(0)
        else:
            return None

    def size(self):
        return len(self.items)


def print_n(queue, n):
    if n < 1 or n > queue.size():
        print("Некоректне значення для n.")
        return

    for i in range(n):
        print(queue.dequeue())


my_queue = Queue()
my_queue.enqueue(1)
my_queue.enqueue(2)
my_queue.enqueue(3)
my_queue.enqueue(4)
my_queue.enqueue(5)

print("Елементи черги:")
print_n(my_queue, 3)
```
